```python
ax = sns.countplot(data = fh1 , x = 'gender' ,palette = ['blue','red'])

for patch in ax.patches:
    h = patch.get_height()
    w = patch.get_x()
    print(h)
    print(w)
    ax.annotate(h,xy = ((w+ 0.3),(h/2)))
```
